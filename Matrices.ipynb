```python
print("Bienvenido a calcula tus matrices!")
print()
print("Dimensiones de la primera Matriz")
fila1= -1
while fila1 < 0:
    fila1 = int(input('Numero de filas: '))
    if fila1 < 0:
        print('Debe ser un numero positivo')
columna1=-1 
while columna1 < 0:
    columna1 = int(input('Numero de columnas'))
    if columna1 < 0:
        print('Debe ser un numero positivo')
print()
print("Dimensiones de la segunda Matriz")
fila2=columna1
print("Numero de filas: ",fila2,"        (Recordatorio: Debe ser igual al numero de columnas de la Matriz 1!)")
columna2= -1
while columna2 < 0:
    columna2 = int(input('Numero de columnas: '))
    if columna2 < 0:
        print('Debe ser un numero positivo')
print()
print()
print("Ingrese los elementos de la Matriz 1")
matriz1=[[]]

for i in range(fila1):
    
    for t in range(columna1):
        matriz1[i].append(int(input("Elemento "+str(t+1)+" de la fila "+str(i+1)+": ")))
    if i<fila1-1:
        matriz1.append([])
print("Matriz 1:")
print(matriz1)
print()

print("Ingrese los elementos de la Matriz 2")
matriz2=[[]]

for i in range(fila2):
    
    for t in range(columna2):
        matriz2[i].append(int(input("Elemento "+str(t+1)+" de la fila "+str(i+1)+": ")))
    if i<fila2-1:
        matriz2.append([])
print("Matriz 2:")
print(matriz2)
print()
print()
matrizR=[[]]
for i in range(fila1):
    for t in range (columna2):
        aux=0
        for s in range(fila2):
            aux += matriz1[i][s]*matriz2[s][t]
        matrizR[i].append(aux) 
    if i<fila1-1:
        matrizR.append([])    

print("Matriz Resultante:")
print(matrizR)
```
